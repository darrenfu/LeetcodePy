```python
# TLE for recursion
from typing import List
from functools import lru_cache
class Solution:
    def probabilityOfHeads(self, prob: List[float], target: int) -> float:
        @lru_cache(None)
        def dp(start: int, target: int) -> float:
            if start == len(prob) - 1:
                if target == 1:
                    return prob[start]
                elif target == 0:
                    return 1.0-prob[start]
                else:
                    return 0.0
            return (1.0-prob[start]) * dp(start+1, target) + prob[start] * dp(start+1, target-1)
        return dp(0, target)
```
