```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.children = dict()
        
    #def __eq__(self, other):
    #    if isinstance(other, str):
    #        return self.path == other
    #    return NotImplemented
        
class FileSystem:
    def __init__(self):
        self.root = TreeNode(-1)

    def create(self, path: str, value: int) -> bool:
        if path == "/" or path == "": return False
        nodes = self.root.children
        dirs = path.strip("/").split("/")
        L = len(dirs)
        for i, d in enumerate(dirs):
            if d == "": return False
            if i == L-1:
                if d in nodes: return False  # dir exist
                nodes[d] = TreeNode(d, value)
            else:
                if d not in nodes: return False  # parent not exist
                nodes = nodes[d].children
        return True

    def get(self, path: str) -> int:
        if path == "/" or path == "": return -1
        dirs = path.strip("/").split("/")
        nodes = self.root.children
        v = -1
        for d in dirs:
            if d not in nodes: return -1
            v = nodes[d].val
            nodes = nodes[d].children
        return v
        


# Your FileSystem object will be instantiated and called as such:
# obj = FileSystem()
# param_1 = obj.create(path,value)
# param_2 = obj.get(path)
```

```python
# from @yihuikang  

class FS:
    def __init__(self, val):
        self.folders = {}
        self.val = val
    
class FileSystem:
    def __init__(self):
        self.root = FS(-1)
        

    def create(self, path: str, value: int) -> bool:
        w = path.split("/")
        n = len(w)
        p = self.root
        # check non leaf directories
        for i in range(1, n-1):
            x = w[i]
            if x not in p.folders:
                return False
            p = p.folders[x]
        # check leaf directory
        if w[-1] in p.folders:
            return False
        p.folders[w[-1]] = FS(value)
        return True
        
        
    def get(self, path: str) -> int:
        w = path.split("/")
        n = len(w)
        p = self.root
        for i in range(1, n):
            x = w[i]
            if x not in p.folders:
                return -1
            p = p.folders[x]
        return p.val
        


# Your FileSystem object will be instantiated and called as such:
# obj = FileSystem()
# param_1 = obj.create(path,value)
# param_2 = obj.get(path)
```
