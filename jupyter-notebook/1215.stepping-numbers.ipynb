```python
#* @awice
# DFS from unit digits, e.g. for 0, 0->1->10,12->101,121,123->...
# compare unit digit of number with last digit of the remainer
# e.g. 1234: (12)3 vs 4
from typing import List
class Solution:
    def countSteppingNumbers(self, low: int, high: int) -> List[int]:
        def search(n: int) -> None:
            if n > high: return
            if n >= low: 
#                 res.append(n)
                res.add(n)
            unitDigit = n % 10 # unit digit
            for newUnitDigit in (unitDigit-1, unitDigit+1):
                if 0 <= newUnitDigit < 10:
                    search(n * 10 + newUnitDigit)
        res = set() # to dedupe duplicates introduced by '0' branch
        # because 0->1, but other digits will lead to two-digit number in first propogation (e.g. 1->10, 1->12)
#         res = []
        for v in range(10): # start with 0~9
            search(v)
#         return res
        return sorted(res)
```
