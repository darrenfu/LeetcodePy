### [348\. Design Tic-Tac-Toe](https://leetcode.com/problems/design-tic-tac-toe/)

Difficulty: **Medium**


Design a Tic-tac-toe game that is played between two players on a _n_ x _n_ grid.

You may assume the following rules:

1.  A move is guaranteed to be valid and is placed on an empty block.
2.  Once a winning condition is reached, no more moves is allowed.
3.  A player who succeeds in placing _n_ of their marks in a horizontal, vertical, or diagonal row wins the game.

**Example:**  

```
Given n = 3, assume that player 1 is "X" and player 2 is "O" in the board.

TicTacToe toe = new TicTacToe(3);

toe.move(0, 0, 1); -> Returns 0 (no one wins)
|X| | |
| | | |    // Player 1 makes a move at (0, 0).
| | | |

toe.move(0, 2, 2); -> Returns 0 (no one wins)
|X| |O|
| | | |    // Player 2 makes a move at (0, 2).
| | | |

toe.move(2, 2, 1); -> Returns 0 (no one wins)
|X| |O|
| | | |    // Player 1 makes a move at (2, 2).
| | |X|

toe.move(1, 1, 2); -> Returns 0 (no one wins)
|X| |O|
| |O| |    // Player 2 makes a move at (1, 1).
| | |X|

toe.move(2, 0, 1); -> Returns 0 (no one wins)
|X| |O|
| |O| |    // Player 1 makes a move at (2, 0).
|X| |X|

toe.move(1, 0, 2); -> Returns 0 (no one wins)
|X| |O|
|O|O| |    // Player 2 makes a move at (1, 0).
|X| |X|

toe.move(2, 1, 1); -> Returns 1 (player 1 wins)
|X| |O|
|O|O| |    // Player 1 makes a move at (2, 1).
|X|X|X|
```

**Follow up:**  
Could you do better than O(_n_<sup>2</sup>) per `move()` operation?

````python
from typing import List

class TicTacToe:

    def __init__(self, n: int):
        """
        Initialize your data structure here.
        """
        self.H = [[0,0] for _ in range(n)]
        self.V = [[0,0] for _ in range(n)]
        self.D = [[0,0] for _ in range(2)]
        self.n = n
        

    def move(self, row: int, col: int, player: int) -> int:
        """
        Player {player} makes a move at ({row}, {col}).
        @param row The row of the board.
        @param col The column of the board.
        @param player The player, can be either 1 or 2.
        @return The current winning condition, can be either:
                0: No one wins.
                1: Player 1 wins.
                2: Player 2 wins.
        """
        def isWon(lines: List[List[int]], m: int, n: int) -> bool:
            return lines[m][n] == self.n
        pid = player-1
        self.H[row][pid] += 1
        if isWon(self.H, row, pid):
            return player
        self.V[col][pid] += 1
        if isWon(self.V, col, pid):
            return player
        if row == col: # diagonal
            self.D[0][pid] += 1 
            if isWon(self.D, 0, pid):
                return player
        if row + col == self.n-1: # back-diagonal
            self.D[1][pid] += 1
            if isWon(self.D, 1, pid):
                return player
        return 0
````

````python
# https://leetcode.com/problems/design-tic-tac-toe/discuss/81932/Python-13-lines-easy-to-understand
class TicTacToe:

    def __init__(self, n: int):
        self.H = [0] * n
        self.V = [0] * n
        self.D = [0,0]
        self.n = n

    def move(self, row: int, col: int, player: int) -> int:
        # counting offset can save one dimension (playerid) in init data structure
        offset = player * 2 - 3 # player 1 -> -1; player 2 -> 1
        self.H[row] += offset
        self.V[col] += offset
        if row == col: # diagonal
            self.D[0] += offset
        if row + col == self.n - 1: # back diagonal
            self.D[1] += offset
        # winning condition met: -n or n
        if offset * self.n in (self.H[row], self.V[col], self.D[0], self.D[1]):
            return player
        return 0
````
